# Load MalwareBazaar embeddings

```python
from tqdm import tqdm
import torch
malware_dataset = torch.load('../datasets/embeddings/embeddings_dict.pt')
```

```python
import requests
import json
import pandas as pd
from tqdm import tqdm
api = 'https://mb-api.abuse.ch/api/v1/'
malware_df = pd.DataFrame(columns=['sha256_hash', 'sha3_384_hash', 'sha1_hash', 'md5_hash', 'first_seen', 'last_seen', 'file_name', 'file_size', 'file_type_mime', 'file_type', 'reporter', 'origin_country', 'anonymous', 'signature', 'imphash', 'tlsh', 'telfhash', 'ssdeep', 'dhash_icon', 'comment', 'tags', 'code_sign', 'delivery_method', 'intelligence', 'file_information', 'ole_information', 'yara_rules', 'vendor_intel', 'comments'])

for hash in tqdm(list(malware_dataset.keys())):
    post_payload = {"query":"get_info","hash":hash}
    info = requests.post(api,post_payload)
    res = json.loads(info.text)
    # malware_df = malware_df.append(pd.DataFrame.from_dict(dict(res['data'][0])),orient='index')
    # print(res['data'][0].keys())
    dictino = {}
    try:
        for key in res['data'][0].keys():
            dictino[key] = str(res['data'][0][key])

        malware_df = malware_df.append(pd.DataFrame([dictino]))
    except:
        continue

    # print(res['data'][0]['signature'])
malware_df['signature']
```

```python
malware_df.to_csv('../datasets/embeddings/malwarebazaar_description.csv', index=False)
```

# Load Pandas Dataframes

```python
import pandas as pd
exploit_db_cve = pd.read_csv('../datasets/exploit_db_cve.csv')
```

```python
exploit_db_cve
```

```python
import requests
from bs4 import BeautifulSoup
nist = 'https://nvd.nist.gov/vuln/detail/'
dataset_cve = {}
for key in dataset:
    cve = exploit_db_cve.loc[exploit_db_cve['exploit_db_id'] == key]
    cves = str(cve.iloc[0]['cve_id']).split(' ')
    cves = list(filter(None, cves))
    dataset_cve[key] = {}
    for c in cves:
        try:
            page = requests.get(nist + c)
            soup = BeautifulSoup(page.text,'html.parser')
            desc = soup.find("p", {"data-testid":"vuln-description"}).text
            dataset_cve[key][c] = desc
        except:
            dataset_cve[key][c] = 'No description field in ' + nist + c

    
    #print(str(key) + ' | ' + str(cve.iloc[0]['cve_id']))
```

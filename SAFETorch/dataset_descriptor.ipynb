# Load Dataset

```python
from tqdm import tqdm
import torch
dataset = torch.load('../datasets/exploits_embeddings_complete.pt')
```

# Load Pandas Dataframes

```python
import pandas as pd
exploit_db_cve = pd.read_csv('../datasets/exploit_db_cve.csv')
```

```python
import requests
from bs4 import BeautifulSoup
nist = 'https://nvd.nist.gov/vuln/detail/'
dataset_cve = {}
for key in dataset:
    cve = exploit_db_cve.loc[exploit_db_cve['exploit_db_id'] == key]
    cves = str(cve.iloc[0]['cve_id']).split(' ')
    cves = list(filter(None, cves))
    for c in cves:
        page = requests.get(nist + c)
        soup = BeautifulSoup(page.text,'html.parser')
        print(soup.find("p", {"data-testid":"vuln-description"}))
    #dataset_cve[key] = list(filter(None, dataset_cve[key]))
    #print(str(key) + ' | ' + str(cve.iloc[0]['cve_id']))
```
